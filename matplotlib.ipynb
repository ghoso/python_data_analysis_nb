```python
# サブプロット
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
_ = ax1.hist(np.random.randn(100),bins=20, color='w',alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(),'w--')
```

```python
# マーカー
from numpy.random import randn
plt.plot(randn(30).cumsum(),"wo--")
```

```python
# 凡例、ラベル
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
# X座標目盛設定
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],
                           rotation=30, fontsize='small')
# タイトル
ax.set_title('My first matplotlib plot')
# x軸のラベル
ax.set_xlabel('Stages')
ax.set_ylabel('Level')

ax.plot(np.random.randn(1000).cumsum())
```

```python
# 複数グラフの凡例
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'w', label='one')
ax.plot(np.random.randn(1000).cumsum(),'r', label='two')
ax.plot(np.random.randn(1000).cumsum(),'b', label='three')
ax.legend(loc='best')
```

```python
# Pandas グラフメソッド
# Seriesデータでグラフ表示
import pandas as pd
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0,100,10))
s.plot()
```

```python
# DataFrameデータでグラフ表示
# 各列のデータを折れ線グラフで表示
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                 columns = ['A','B','C','D'],
                  index=np.arange(0,100,10)) 
df
```

```python
df.plot()
```

```python
# Seaborn
import seaborn as sns

# データ読み込み
tips = pd.read_csv('tips.csv')
# チップ割合計算
tips['tip_pct'] = tips['tip']/(tips['total_bill'] - tips['tip'])
tips.head()
```

```python
# 棒グラフ
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
# 散布図
sns.regplot('total_bill', 'tip', data=tips)
```

```python
# 曜日ごとのチップ割合（時間帯別、喫煙/非喫煙別）
sns.catplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```
